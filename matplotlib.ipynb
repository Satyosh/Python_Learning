# 折れ線グラフ

折れ線グラフを描くときはplot関数を使う。

plotの第1引数にX軸側の値のリスト、第２引数にY軸側の値のリストを指定する。

```python
plt.rcParams['font.size'] = 14

# グラフを描画するキャンバスの用意
fig, ax = plt.subplots(facecolor="white")

# 折れ線グラフはplot関数を使う
ax.plot(["月","火","水","木","金","土","日"],[19,22,24,21,20,22,20])
ax.set_xlabel('曜日')
ax.set_ylabel('気温',rotation="vertical")
ax.set_yticks(range(0,31,5))
ax.set_title('東京の気温')
ax.grid()
ax.legend()
fig.savefig("./graph1.png")
```

```python
ax.bar(["月","火","水","木","金","土","日"],[19,22,24,21,20,22,20],label="最高気温",color='r')
ax.plot(["月","火","水","木","金","土","日"],[10,13,12,12,10,11,12],label='最低気温',marker='x',color='b')
ax.set_xlabel('曜日')
ax.set_ylabel('気温',rotation="vertical")
ax.set_yticks(range(0,31,5))
ax.set_title('東京の気温')
ax.grid()
ax.legend()
```

```python
np.random.seed(40)

len = np.random.uniform(140,170,size=30).round(1)
bmi = np.random.uniform(20,24,size=30).round(2)
wt = (bmi*(len/100)**2)

fig, ax = plt.subplots()

ax.scatter(len,wt)
ax.set_xlabel('身長')
ax.set_ylabel('体重')
```

```python
fig, ax = plt.subplots()

ax.hist(len,bin=10)
ax.set_title('なんか')
```

```python
# 複数のグラフ

fig, ax = plt.subplots(2,2,figsize=(15,8))

ax[0,0].plot(["月","火","水","木","金","土","日"],[10,13,12,12,10,11,12],label='最低気温',marker='x',color='b')
ax[0,0].set_title('[0,0]のグラフです')

ax[0,1].plot(["月","火","水","木","金","土","日"],[19,22,24,21,20,22,20],label="最高気温",marker='o')
ax[0,1].set_title('[0,1]のグラフです')

ax[1,0].scatter(len,wt)
ax[1,1].hist(wt)
```
